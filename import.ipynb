**Importing libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import sqlite3
```

SQLite  file is a database engine (complex binanry format) packed in a file that python can't read directly.
So we must connect first ie talk to the engine: open a conversation (con=sqlite.connect) and then get a messenger : (con.cursor) and send SQL queries (cursor.excute)

```python
conn = sqlite3.connect("olist.sqlite")
Tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type = 'table'",conn)
print(Tables)
Pers = pd.read_sql_query("SELECT * FROM products",conn)
print(Pers)
Pers.head()
```
